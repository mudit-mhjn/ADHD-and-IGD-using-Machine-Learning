```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
```

```python
df = pd.read_csv('dataset.csv')
df
```

Drop irrelevant features or features wih negetive corelation value.<br>
Then we split the dataset into Test and Train for training and testing of model.<br>
Standardization of features in the dataset before training.<br>
Training the Model<br>
Calculation of accuracy on test and train set.<br>

```python
train = df.drop(labels=['no_of_calls', 'call_duration', 'uti'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train.drop('result',axis=1), 
                                                    train['result'], test_size=0.20, random_state = 101)
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
X_train = std.fit_transform(X_train)
X_test = std.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB

clf = GaussianNB()
clf.fit(X_train, y_train)
target_pred = clf.predict(X_test)
```

```python
print(target_pred)

print("Accuracy:",metrics.accuracy_score(y_test, target_pred))
```

```python
train_pred = clf.predict(X_train)
print(metrics.accuracy_score(y_train, train_pred))
#Accuracy on train
```

```python
metrics.plot_confusion_matrix(clf, X_test, y_test, display_labels=['Negative', 'Positive'])
```
